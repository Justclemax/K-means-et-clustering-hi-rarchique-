2. Utilisez le clustering hiérarchique pour identifier les regroupements inhérents au sein de vos données.

```python
from sklearn.cluster import AgglomerativeClustering

# Sélectionnez les caractéristiques à utiliser pour le clustering
X = df[["BALANCE", "PURCHASES","CASH_ADVANCE"]]

# Effectuez le clustering hiérarchique
clustering = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete').fit(X)

# Obtenez les étiquettes de cluster pour chaque échantillon
clust_labels=clustering.fit_predict(X)  # Mehode 1 


# Ajoutez les prédictions de cluster à notre jeu de données
df['cluster'] = clust_labels

# Affichez le nombre de points de données dans chaque cluster
df['cluster'].value_counts()

```

```python
data=df
data.head()
```

3. Tracez les clusters. 

```python


import matplotlib.pyplot as plt

# Sélectionnez les colonnes à utiliser pour tracer les clusters
x_axis = data['BALANCE']
y_axis = data['PURCHASES']

# Créez une figure et un sous-graphique
fig, ax = plt.subplots()

# Utilisez un boucle for pour dessiner des points de chaque cluster avec une couleur différente
colors = {0:'red', 1:'blue', 2:'green'}
for cluster_num, color in colors.items():
    x = x_axis[clust_labels == cluster_num]
    y = y_axis[clust_labels == cluster_num]
    ax.scatter(x, y, c=color, label=cluster_num, alpha=0.5)

# Ajoutez un titre et des étiquettes aux axes
ax.set_title('Clustering hiérarchique des données de carte de crédit')
ax.set_xlabel('Solde de la carte de crédit')
ax.set_ylabel('Achats')
```
